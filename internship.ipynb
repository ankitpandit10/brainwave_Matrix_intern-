```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("sales_06_FY2020-21 copy.csv")

# Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# --- 1. Sales Performance ---
# Monthly Sales Trend
monthly_sales = df.groupby(df['order_date'].dt.to_period('M'))['total'].sum().reset_index()
monthly_sales['order_date'] = monthly_sales['order_date'].astype(str)
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='order_date', y='total')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_categories = df.groupby('category')['total'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar')
plt.title('Top-Selling Categories')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Sales by Region
region_sales = df.groupby('Region')['total'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
region_sales.plot(kind='bar', color='skyblue')
plt.title('Sales by Region')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
# --- 2. Customer Segmentation ---
# Age Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Customer Age Distribution')
plt.show()
```

```python
# Gender-wise Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Gender')
plt.title('Customer Gender Distribution')
plt.show()
```

```python
# --- 3. Discount Impact ---
# Discount vs. Quantity Ordered
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Discount_Percent', y='qty_ordered', alpha=0.3)
plt.title('Discount Percent vs Quantity Ordered')
plt.show()
```

```python
# Total Discount Given Over Time
df['discount_value'] = df['value'] * df['Discount_Percent'] / 100
monthly_discount = df.groupby(df['order_date'].dt.to_period('M'))['discount_value'].sum().reset_index()
monthly_discount['order_date'] = monthly_discount['order_date'].astype(str)
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_discount, x='order_date', y='discount_value')
plt.title('Monthly Discount Given')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- 4. Order & Product Insights ---
# Average Order Value
avg_order_value = df.groupby('order_id')['total'].sum().mean()
print(f"Average Order Value: ${avg_order_value:.2f}")
```

```python
# SKU-Level Performance
top_skus = df.groupby('sku')['total'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
top_skus.plot(kind='bar')
plt.title('Top Performing SKUs')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- 5. Payment Method Trends ---
payment_method_counts = df['payment_method'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=payment_method_counts.index, y=payment_method_counts.values)
plt.title('Payment Method Distribution')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- 6. Customer Retention & Loyalty ---
df['Customer Since'] = pd.to_datetime(df['Customer Since'])
df['customer_lifetime_days'] = (df['order_date'] - df['Customer Since']).dt.days
plt.figure(figsize=(10, 6))
sns.histplot(df['customer_lifetime_days'], bins=30, kde=True)
plt.title('Customer Lifetime (Days)')
plt.show()
```

```python
# High-Value Customers
customer_spend = df.groupby('cust_id')['total'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
customer_spend.plot(kind='bar')
plt.title('Top 10 High-Value Customers')
plt.ylabel('Total Spend')
plt.tight_layout()
plt.show()
```
